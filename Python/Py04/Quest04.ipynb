```python
from typing import Callable
##### NEW CALCULATOR #####
# INPUT: '연산할 정수 2개를 입력하세요',
# 연산자(+,-,/,*,**)
# INPUT 으로 정수를 못받는 경우 -> ValueError 처리 -> 오류메세지 '잘못된 입력입니다. 정수를 입력해주세요' -> 다시 프로그램 실행
# INPUT 으로 연산자가 해당되지 않는 경우 -> 오류메세지 '+, -, /, *, ** 에 해당하는 연산자를 입력하세요' -> 다시 프로그램 실행

# 제대로 입력받은 경우 출력값을 주고
# 계속 계산할 것인지 출력 '계속 계산하시겠습니까: Y[Yes], N[No] 를 선택하세요' -> Y면 프로그램 실행. N 이면 프로그램 종료


import math as m
print('======= NEW CALCULATOR =======')
print('이 계산기는 정수 2개와 연산자(+,-,/,*,**)을 입력받습니다')
print('+: 덧셈, -: 뺄셈, /:나누기, *: 곱하기, **:power')
print('==============================')

x=None #첫번째 변수x
y=None #두번째 변수y
cal="" #실행할 연산자 cal
result=None #실행한 결과값을 받을 변수 result
r=1 #프로그램을 실행(run) 시키는 변수. 1일 경우 프로그램을 계속 실행. 0일 경우 프로그램이 종료된다

cal_list=['+', '-', '*', '/', '**'] #이 프로그램에서 실행가능한 연산자들의 리스트

def inputx():
#x를 입력받고, 프로그램에서 제공하지 않는 정수외의 type 일 경우 다시 입력받는 함수
  global x
  try:
    #input 으로 정수 x,y를 입력받는다
     x=int(input((f"첫번째 정수: ")))
  except ValueError:
    #valueError가 뜨는 경우 에러메세지를 출력하고 다시 함수를 돌린다
    print(f"잘못된 입력입니다. 정수를 입력해주세요")
    inputx()

def inputy():
#y를 입력받고, 프로그램에서 제공하지 않는 정수외의 type 일 경우 다시 입력받는 함수
  global y
  try:
    #input 으로 정수y를 입력받는다
    y=int(input("두번째 정수: "))
  except ValueError:
    #valueError가 뜨는 경우 에러메세지를 출력하고 다시 함수를 돌린다
    print(f"잘못된 입력입니다. 정수를 입력해주세요")
    inputy()

def inputcal():
#cal을 입력받고, 프로그램에서 제공하지 않는 연산자일 경우 연산자를 다시 입력받는 함수
  global cal
  cal=input('시행할 연산을 입력하세요')
  if cal not in cal_list :
    #입력받은 cal 이 프로그램에서 실행가능한 연산자에 없는 경우
    #print('a' in cal_list)
    raise Exception("+, -, /, *, ** 에 해당하는 연산자를 입력하세요")


def calculator():
  global result
  #연산자를 실행해주고 결과값을 프린트해주는 함수
  if cal=='+':
    result = x+y
    return print(f'{x}{cal}{y}={result}입니다')
  elif cal=='-':
    result = x-y
    return print(f'{x}{cal}{y}={result}입니다')
  elif cal=='/':
    try:
      result = x/y
      return print(f'{x}{cal}{y}={result}입니다')
    except ZeroDivisionError:
      return print('0으로 나눌수는 없습니다')
  elif cal=='*':
    result = x*y
    return print(f'{x}{cal}{y}={result}입니다')
  elif cal=='**':
    result=m.pow(x,y)
    return print(f'{x}{cal}{y}={result}입니다')


def re_cal():
  #프로그램을 계속실행할지(r=1) 종료할지(r=0) 결정하는 함수
  global r
  try:
    answer=input(print(f'계속 계산하시겠습니까: Y[Yes], N[No] 를 입력하세요'))
    #answer를 입력받고 해당 결과에 따라 r을 1또는 0으로 변경해준다
    if answer=='Y' or answer=='y':
      r=1
    elif answer=='N' or answer =='n':
      r=0
  except:
    #제대로 입력되지 않을 경우 다시 입력하도록 함수를 다시돌린다
    print(f'읽을 수 없습니다. 다시 입력하세요')
    re_cal()

##### CALCULATOR PROGRAM PROCESS #####


while r==1:
  # 프로그램을 유지하는 상태를 나타내는 변수 r 이 1일 때는 프로그램을 계속 실행한다
  inputx()
  inputy()
  inputcal()
  calculator()
  re_cal()
else:
  #프로그램을 유지하자는 상태가 변경될 경우 프로그램을 종료한다.
  print(f'프로그램을 종료합니다')

```

```python
def calculator():
    while True:
        # 첫 번째 정수 입력
        while True:
            try:
                num1 = int(input("첫 번째 정수를 입력하세요: "))
                break  # 올바른 입력이면 반복문을 탈출
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        # 연산자 입력
        while True:
            op = input("연산자를 입력하세요 (+, -, *, /, **): ")
            if op in ['+', '-', '*', '/', '**']:
                break  # 올바른 입력이면 반복문을 탈출
            else:
                print("지원하지 않는 연산자입니다. 올바른 연산자를 입력해주세요.")

        # 두 번째 정수 입력
        while True:
            try:
                num2 = int(input("두 번째 정수를 입력하세요: "))
                break  # 올바른 입력이면 반복문을 탈출
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        # 연산 수행
        try:
            if op == '+':
                result = num1 + num2
            elif op == '-':
                result = num1 - num2
            elif op == '*':
                result = num1 * num2
            elif op == '/':
                result = num1 / num2
            elif op == '**':
                result = num1 ** num2

            print(f"결과: {result}")

        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")

        # 계속할지 여부를 명확히 확인
        while True:
            cont = input("계속 계산하시겠습니까? (y/n): ").strip().lower()
            if cont == 'y':
                break
            elif cont == 'n':
                print("계산기를 종료합니다.")
                return
            else:
                while True:
                    confirm = input("(y/n)으로 확실히 대답하세요, 알겠나요? (y): ").strip().lower()
                    if confirm != 'y':
                        while True:
                            a = 1
                            confirm = input(f"대답{'!' * a} (y): ").strip().lower()
                            a += 1
                            if confirm == 'y':
                                break
                            continue
                        break
                continue

# 함수 실행
calculator()
```
